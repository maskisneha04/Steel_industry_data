```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load and clean data function
def load_and_clean_data(df):
    # Convert 'date' column to datetime format
    df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y %H:%M')

    # Check and handle missing values
    missing_values = df.isnull().sum()
    print("Missing values in each column:\n", missing_values)

    # Drop or fill missing values (example using drop)
    df = df.dropna()
    return df

# Display summary statistics
def display_summary_statistics(df):
    print("Summary statistics for numerical columns:\n", df.describe())
    categorical_columns = ['WeekStatus', 'Day_of_week', 'Load_Type']
    for col in categorical_columns:
        if col in df.columns:
            print(f"\nValue counts for {col}:\n", df[col].value_counts())
        else:
            print(f"\nColumn '{col}' not found in the dataset.")

# Visualization functions, with samples
def plot_usage_distribution(df):
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Usage_kWh'], bins=30, kde=True, color='skyblue')
    plt.title('Distribution of Usage (kWh)')
    plt.xlabel('Usage (kWh)')
    plt.ylabel('Frequency')
    plt.show()

def plot_usage_by_load_type(df):
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x='Load_Type', y='Usage_kWh', palette='Set2')
    plt.title('Usage (kWh) by Load Type')
    plt.xlabel('Load Type')
    plt.ylabel('Usage (kWh)')
    plt.show()

def plot_usage_over_time(df):
    df_sample = df.sample(frac=0.1, random_state=42)  # Use a 10% sample for line plot
    plt.figure(figsize=(14, 6))
    sns.lineplot(data=df_sample, x='date', y='Usage_kWh', color='blue')
    plt.title('Usage (kWh) Over Time (Sampled)')
    plt.xlabel('Date')
    plt.ylabel('Usage (kWh)')
    plt.show()

def plot_week_status_distribution(df):
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x='WeekStatus', hue='Day_of_week', palette='viridis')
    plt.title('Distribution of Week Status by Day of the Week')
    plt.xlabel('Week Status')
    plt.ylabel('Count')
    plt.show()

def plot_correlation_heatmap(df):
    df_sample = df.sample(frac=0.1, random_state=42)  # Use 10% sample for correlation
    corr_matrix = df_sample.select_dtypes(include=['number']).corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu')
    plt.title("Pearson Correlation Heatmap (Sampled)", fontsize=15, color='b', pad=12, loc='center')
    plt.show()

def plot_usage_vs_reactive_power(df):
    df_sample = df.sample(frac=0.1, random_state=42)  # 10% sample for scatter plot
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df_sample, x='Usage_kWh', y='Lagging_Current_Reactive_Power_kVarh', hue='Load_Type', palette='coolwarm')
    plt.title('Usage (kWh) vs. Lagging Current Reactive Power (kVarh) (Sampled)')
    plt.xlabel('Usage (kWh)')
    plt.ylabel('Lagging Current Reactive Power (kVarh)')
    plt.legend(title='Load Type')
    plt.show()

# Main function to execute EDA on a smaller section of the data
def perform_eda(df):
    # Step 1: Display summary statistics
    display_summary_statistics(df)

    # Step 2: Visualizations
    plot_usage_distribution(df)
    plot_usage_by_load_type(df)
    plot_usage_over_time(df)
    plot_week_status_distribution(df)
    plot_correlation_heatmap(df)
    plot_usage_vs_reactive_power(df)

# Usage example
# Assuming 'df' is your DataFrame
df = load_and_clean_data(df)  # Load and clean data
perform_eda(df)  # Perform EDA on sampled data
```

```python
# Additional imports for time series decomposition
from statsmodels.tsa.seasonal import seasonal_decompose

def plot_hourly_trend(df):
    """
    Plots the average Usage_kWh by hour of the day.
    """
    df['hour'] = df['date'].dt.hour
    hourly_usage = df.groupby('hour')['Usage_kWh'].mean()

    plt.figure(figsize=(10, 6))
    sns.lineplot(x=hourly_usage.index, y=hourly_usage.values, color='orange')
    plt.title('Average Usage (kWh) by Hour of Day')
    plt.xlabel('Hour of Day')
    plt.ylabel('Average Usage (kWh)')
    plt.xticks(range(0, 24))
    plt.show()

def plot_daily_trend(df):
    """
    Plots the average Usage_kWh by day of the week.
    """
    daily_usage = df.groupby('Day_of_week')['Usage_kWh'].mean().reindex(
        ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    )

    plt.figure(figsize=(10, 6))
    sns.barplot(x=daily_usage.index, y=daily_usage.values, palette='coolwarm')
    plt.title('Average Usage (kWh) by Day of Week')
    plt.xlabel('Day of Week')
    plt.ylabel('Average Usage (kWh)')
    plt.show()

def plot_power_factor_distribution(df):
    """
    Plots histograms for Lagging and Leading Current Power Factors.
    """
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.histplot(df['Lagging_Current_Power_Factor'], bins=30, color='purple', kde=True)
    plt.title('Distribution of Lagging Current Power Factor')

    plt.subplot(1, 2, 2)
    sns.histplot(df['Leading_Current_Power_Factor'], bins=30, color='green', kde=True)
    plt.title('Distribution of Leading Current Power Factor')
    plt.show()

def plot_load_type_distribution(df):
    """
    Plots the count of Load_Type across WeekStatus and Day_of_week.
    """
    plt.figure(figsize=(14, 6))
    sns.countplot(data=df, x='Load_Type', hue='WeekStatus', palette='Set1')
    plt.title('Load Type Distribution by Week Status')
    plt.xlabel('Load Type')
    plt.ylabel('Count')
    plt.legend(title='Week Status')
    plt.show()

    plt.figure(figsize=(14, 6))
    sns.countplot(data=df, x='Load_Type', hue='Day_of_week', palette='Set2')
    plt.title('Load Type Distribution by Day of Week')
    plt.xlabel('Load Type')
    plt.ylabel('Count')
    plt.legend(title='Day of Week')
    plt.show()

def plot_outliers(df):
    """
    Plots boxplots for Usage_kWh and Lagging_Current_Reactive_Power_kVarh to identify outliers.
    """
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.boxplot(data=df, y='Usage_kWh', color='coral')
    plt.title('Outliers in Usage (kWh)')

    plt.subplot(1, 2, 2)
    sns.boxplot(data=df, y='Lagging_Current_Reactive_Power_kVarh', color='lightblue')
    plt.title('Outliers in Lagging Current Reactive Power (kVarh)')
    plt.show()

# Update main EDA function to include new analyses
def perform_eda(df):
    # Step 4: Additional Analyses
    plot_hourly_trend(df)
    plot_daily_trend(df)
    plot_power_factor_distribution(df)
    plot_load_type_distribution(df)
    plot_outliers(df)

# Run EDA with extended analysis
perform_eda(df)
```
